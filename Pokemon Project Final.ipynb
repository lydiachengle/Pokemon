Pokémon types dictate whether one Pokémon's attack is strong or weak (i.e., super effective or not very effective) against another, dealing additional or reduced damage as a result. What is the modifier in the damage formula above? Our simplified modifier will be:

$$ Modifier = Critical \times Random \times Type $$

**Random** is a random integer percentage between 0.85 and 1.00 (inclusive).

Recall that **Type** can be 0.5 (not very effective), 1 (normally effective), or 2 (super effective), depending on how effective the move's type is against the defending Pokémon's type.

Actually, the type can also take on the values ¼ and 4. For targets that have two types (every Pokémon is either one or two types ), such as Bulbasaur, who is both a Grass and Poison type, overall type effectiveness is the product of the effectiveness against each of its types:

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;-If the type of a move is super effective against both of the opponent's types (such as Dig, a Ground-type move, used against an Aggron, a Steel/Rock Pokémon), then the move does 2x2=4 times the damage;

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;-If the type of a move is not very effective against both of the opponent's types (such as Wake-Up Slap, a Fighting-type move, used against a Sigilyph, a Psychic/Flying Pokémon), then the move only does 1/2x1/2=¼ of the damage;

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;-If the type of a move is super effective against one of the opponent's types but not very effective against the other (such as Razor Leaf, a Grass-type move, used against a Gyarados, a Water/Flying Pokémon), then the move deals normal damage 2x1/2=1;

**Critical** is 2 if you have landed a critical hit and 1 otherwise. What is a critical hit? This is of the one last things to understand. A critical hit is a random event that multiplies the damage of a damage-dealing move. When one occurs, the game will display the message "A critical hit!" after the damage has been dealt. 

How does one calculate the probability that the attack was critical? The probability that an attacking Pokémon landed a critical hit based on that attacking Pokémon's speed is: $$ P = Speed / 512 $$

Thus, one can generate a random  number between 0 and 511 and if that number is less than the attacking Pokémon's speed then Critical=2.

Now let's return to levels. When a Pokémon beats another Pokémon, he gains their experience points. With enough experience points, the Pokémon achieves the next level. Levels go from 1 to 100. When a player just starts out, they have 0 experience points and are on level 1. Once they gain experience points, the function that models experience points versus level is the cube root function:

$$ Level = (Experience)^{1/3} $$

We can now take the level of the attacker into account when calculating damage:

```python
# %%pycodestyle
class Game_Play:
    '''
    This class facilitates game play
    '''

    def best_move(self, computer, player):
        '''
        Takes in the computer's pokémon and the player's
        pokémon; returns dictionary with the key being how
        effective the move's type is against the player's
        pkm and the value being the moves
        with that specific type modifier.
        '''

        #  initializes dict of each type and the types they are
        #  super effective (ie 2x) against
        SUPER_EFFECTIVE = {
            "Normal": ["N/A"],
            "Fire": ["Grass", "Ice", "Bug", "Steel"],
            "Water": ["Fire", "Ground", "Rock"],
            "Electric": ["Water", "Flying"],
            "Grass": ["Water", "Ground", "Rock"],
            "Ice": ["Grass", "Ground", "Flying", "Dragon"],
            "Fighting": ["Normal", "Ice", "Rock", "Dark", "Steel"],
            "Poison": ["Grass", "Fairy"],
            "Ground": ["Fire", "Electric", "Poison", "Rock", "Steel"],
            "Flying": ["Grass", "Fighting", "Bug"],
            "Psychic": ["Fighting", "Poison"],
            "Bug": ["Grass", "Psychic", "Dark"],
            "Rock": ["Fire", "Ice", "Flying", "Bug"],
            "Ghost": ["Psychic", "Ghost"],
            "Dragon": ["Dragon"],
            "Dark": ["Psychic", "Ghost"],
            "Steel": ["Ice", "Rock", "Fairy"],
            "Fairy": ["Fighting", "Dragon", "Dark"]
        }

    # initializes dict of each type and the types they are
    # not very affective (ie 0.5x) against.
        NOT_VERY_EFFECTIVE = {
            "Normal": ["Rock", "Steel"],
            "Fire": ["Fire", "Water", "Rock", "Dragon"],
            "Water": ["Water", "Grass", "Dragon"],
            "Electric": ["Electric", "Grass", "Dragon"],
            "Grass": ["Fire", "Grass", "Poison", "Flying",
                      "Bug", "Dragon", "Steel"],
            "Ice": ["Fire", "Water", "Ice", "Steel"],
            "Fighting": ["Poison", "Flying", "Psychic", "Bug", "Fairy"],
            "Poison": ["Poison", "Ground", "Rock", "Ghost"],
            "Ground": ["Grass", "Bug"],
            "Flying": ["Electric", "Rock", "Steel"],
            "Psychic": ["Psychic", "Steel"],
            "Bug": ["Fire", "Fighting", "Poison", "Flying", "Ghost",
                    "Steel", "Fairy"],
            "Rock": ["Fighting", "Ground", "Steel"],
            "Ghost": ["Dark"],
            "Dragon": ["Steel"],
            "Dark": ["Fighting", "Dark", "Fairy"],
            "Steel": ["Fire", "Water", "Electric", "Steel"],
            "Fairy": ["Fire", "Poison", "Steel"]
        }

    # dictionary of each move (key) and their type (value)
        REV_MOVES = {
            'Scratch': 'Normal',
            'Tackle': 'Normal',
            'Pound': 'Normal',
            'Rage': 'Normal',
            'Fury Attack': 'Normal',
            'Body Slam': 'Normal',
            'Double Slap': 'Normal',
            'Headbutt': 'Normal',
            'Ember': 'Fire',
            'Fire Spin': 'Fire',
            'Bubble': 'Water',
            'Aqua Jet': 'Water',
            'Thunder Shock': 'Electric',
            'Thunderbolt': 'Electric',
            'Vine Whip': 'Grass',
            'Magical Leaf': 'Grass',
            'Razor Leaf': 'Grass',
            'Absorb': 'Grass',
            'Ice Shard': 'Ice',
            'Double Kick': 'Fighting',
            'Smog': 'Poison',
            'Earthquake': 'Ground',
            'Wing Attack': 'Flying',
            'Peck': 'Flying',
            'Confusion': 'Psychic',
            'Dream Eater': 'Psychic',
            'Twineedle': 'Bug',
            'Struggle Bug': 'Bug',
            'Rock Throw': 'Rock',
            'Rock Slide': 'Rock',
            'Lick': 'Ghost',
            'Shadow Ball': 'Ghost',
            'Outrage': 'Dragon',
            'Crunch': 'Dark',
            'Bite': 'Dark',
            'Flash Cannon': 'Steel',
            'Fairy Wind': 'Fairy',
            'Draining Kiss': 'Fairy'
        }

        # initializes dictionary to hold the results
        good_or_bad = {}

        # for each move the computer can do, calculates
        # how effective it will be against the player's
        # pokémon's type(s).
        for move in computer.moves:
            move_type = REV_MOVES[move]

            t_modifier = 1
            for _type in player.type_:
                if _type in SUPER_EFFECTIVE[move_type]:
                    t_modifier *= 2
                elif _type in NOT_VERY_EFFECTIVE[move_type]:
                    t_modifier *= 0.5
                else:
                    t_modifier *= 1

            good_or_bad[t_modifier] = good_or_bad.get(t_modifier, []) + [move]

        return good_or_bad
    
    def move_help(self, p_pkm, o_pkm):
        # Creating a dictionary that contains player's move
        # and its corresponding damage 
        POWERS = {
            'Scratch': 40,
            'Tackle': 40,
            'Pound': 40,
            'Rage': 20,
            'Fury Attack': 15,
            'Ember': 40,
            'Fire Spin': 35,
            'Bubble': 40,
            'Aqua Jet': 40,
            'Thunder Shock': 40,
            'Thunderbolt': 90,
            'Vine Whip': 45,
            'Magical Leaf': 60,
            'Ice Shard': 40,
            'Double Kick': 30,
            'Earthquake': 100,
            'Wing Attack': 60,
            'Peck': 35,
            'Confusion': 50,
            'Twineedle': 25,
            'Rock Throw': 50,
            'Rock Slide': 75,
            'Lick': 30,
            'Outrage': 120,
            'Crunch': 80,
            'Bite': 60,
            'Flash Cannon': 80,
            'Smog': 30,
            'Dream Eater': 100,
            'Body Slam': 85,
            'Double Slap': 15,
            'Razor Leaf': 55,
            'Headbutt': 70,
            'Absorb': 20,
            'Fairy Wind': 40,
            'Struggle Bug': 50,
            'Draining Kiss': 50,
            'Shadow Ball': 80
                }
        bestmove = []
        for type_mod, moves in self.best_move(p_pkm, o_pkm).items():
            for move in moves:
                bestmove.append((POWERS[move]*type_mod, move))
        return sorted(bestmove)[-1]

    def battle(self, defeated, player):
        '''
        simulates a single battle between the player and the computer
        '''

        import random as rand

        # shows the set of all the player's available pokémon
        print(f'Your character choices are {player.ch_list(defeated)}')

        character = input('Which character would you like? ').title()

        # only allows the character to choose a pokémon in the set
        while character not in player.ch_list(defeated):
            print(('You did not input a possible character. '
                  'Please check spelling.'))
            character = input('Which character would you like? ').title()

        # initializes the player's pokémon
        p_pkm = Pokemon(character)

        # list of all the possible pokémon charachters
        pokemon_characters = ['Pikachu', 'Charizard', 'Squirtle',
                              'Jigglypuff', 'Gengar', 'Magnemite',
                              'Bulbasaur', 'Charmander', 'Beedrill',
                              'Golem', 'Dewgong', 'Hypno', 'Cleffa',
                              'Cutiefly']

        # computer chooses a character,
        # then initializes the computer (or opponent) pokémon
        comp_character = rand.choice(pokemon_characters)
        print(f'Your opponent has chosen a {comp_character}.')
        o_pkm = Pokemon(comp_character)

        # prints the starting hit points of both the player and the computer
        print(f'\nStarting HP:\n\tPlayer: {p_pkm.hp}\n\tOpponent: {o_pkm.hp}')

        # determines which pokémon starts based on speed
        if p_pkm.speed >= o_pkm.speed:
            counter = 2
            print('Player starts!')
        else:
            counter = 1
            print('Opponent starts.')

        # Finds the most damaging moves (according to type)
        # from the list generated from selt.best_move
        highest = 0
        for key in self.best_move(o_pkm, p_pkm).keys():
            if key > highest:
                highest = key
                
        sacrifice = input("Would you like to sacrifice 10% "
                          "of your HP in exchange for the best "
                          "move against your opponent? y/n ")
        if sacrifice == "y":
            p_pkm.hp = 0.9 * p_pkm.hp
            print('Your best move is', self.move_help(p_pkm, o_pkm)[-1])
            print("Your current exp: ", p_pkm.hp)


        # the pokémon alternate damaging each other
        while (p_pkm.hp > 0 and o_pkm.hp > 0):
            if counter % 2 == 0:
                print('\nYour turn:')
                print(f'The moves avaliable are {p_pkm.moves}')
                move = input('Which move would you like? ').title()

                # ensures that player only picks a move in the list
                while move not in p_pkm.moves:
                    print(('You did not input a possible move. '
                          'Please try again.'))
                    move = input('Which move would you like ').title()

                # subtracts the damage of the player's pkm from
                # the opponent's hp
                o_pkm.hp -= p_pkm.damage(move, player.level, o_pkm.type_)
                print(f"Your HP: {p_pkm.hp:.2f} Opponent's HP: {o_pkm.hp:.2f}")

            else:
                print("\nOpponent's turn: ")

                # the computer chooses a random move from the
                # list of best moves
                # power does not play a factor because if the computer
                # chose the actual best move 100% of the time the player
                # would not win as much and therefore
                # would not like the game as much.
                # additionally, the computer always uses the player's level
                # so that the computer and player are evenly matched
                p_pkm.hp -= o_pkm.damage(
                    rand.choice(self.best_move(o_pkm, p_pkm)[highest]),
                                player.level, p_pkm.type_)
                print(f"Your HP: {p_pkm.hp:.2f} Opponent's HP: {o_pkm.hp:.2f}")

            counter += 1

        # returns a tuple stating the winner of the game.
        # In the case that the player wins, the program returns
        # the computer pokémon's name and experience points
        if p_pkm.hp >= o_pkm.hp:
            print('Player wins!!')
            return tuple(('player', o_pkm.name, o_pkm.exp))
        elif o_pkm.hp > p_pkm.hp:
            print('Computer wins!!')
            return 'computer'

    def main(self):
        '''
        facilitates actual game play
        '''
        # list of all the possible pokémon charachters
        pokemon_characters = ['Pikachu', 'Charizard', 'Squirtle',
                              'Jigglypuff', 'Gengar', 'Magnemite',
                              'Bulbasaur', 'Charmander', 'Beedrill',
                              'Golem', 'Dewgong', 'Hypno', 'Cleffa',
                              'Cutiefly']
        
        # determines if the player has played Pokémon before,
        # and if so, allows them to start at their previous level
        before = input('Have you played Pokémon before? (y/n) ').lower()
        if before == 'y':
            defeated_input = input(('Which Pokémon did you defeat? '
                             '(separate by comma space) ')).title().split(', ')
            defeated = []
            for pkm in defeated_input:
                if pkm in pokemon_characters:
                    defeated.append(pkm)
            points = int(input('How many exp points did you have? '))
        else:
            points = 0
            defeated = []

        # initialize the player's experience points (and therefore level)
        player = Player(points)

        ask = 'y'

        # creates battles until the player loses or does not want to play
        while ask != 'n':

            result = self.battle(defeated, player)

            if result[0] == 'c':
                print('Game over!')  # stops the game if the player loses
                return None
            elif result[0] == 'player':
                # adds the defeated pokémon to the list of possible characters
                defeated.append(result[1])
                player.exp += result[2]  # updates the player's exp points
                player.update_level()  # updates the player's level

            print((f'You have {player.exp} points, and the Pokémon you '
                  f'have defeated are {set(defeated)}.'))

            ask = input('Would you like to keep playing? (y/n) ').lower()


if __name__ == '__main__':
    Game_Play().main()
```
